```python
import turtle

# Configuração da janela
janela = turtle.Screen()
janela.title("Tênis no Telejogo")
janela.bgcolor("black")
janela.setup(width=600, height=400)

# Raquete esquerda
raquete_esquerda = turtle.Turtle()
raquete_esquerda.speed(0)
raquete_esquerda.shape("square")
raquete_esquerda.color("white")
raquete_esquerda.shapesize(stretch_wid=6, stretch_len=2)
raquete_esquerda.penup()
raquete_esquerda.goto(-250, 0)

# Raquete direita
raquete_direita = turtle.Turtle()
raquete_direita.speed(0)
raquete_direita.shape("square")
raquete_direita.color("white")
raquete_direita.shapesize(stretch_wid=6, stretch_len=2)
raquete_direita.penup()
raquete_direita.goto(250, 0)

# Bola
bola = turtle.Turtle()
bola.speed(40)
bola.shape("circle")
bola.color("white")
bola.penup()
bola.goto(0, 0)
bola.dx = 2
bola.dy = -2

# Placar
placar_esquerda = 0
placar_direita = 0
placar = turtle.Turtle()
placar.speed(0)
placar.color("white")
placar.penup()
placar.hideturtle()
placar.goto(0, 170)
placar.write("Jogador 1: 0  Jogador 2: 0", align="center", font=("Courier", 20, "normal"))

# Funções
def raquete_esquerda_cima():
    y = raquete_esquerda.ycor()
    y += 30
    raquete_esquerda.sety(y)

def raquete_esquerda_baixo():
    y = raquete_esquerda.ycor()
    y -= 30
    raquete_esquerda.sety(y)

def raquete_direita_cima():
    y = raquete_direita.ycor()
    y += 30
    raquete_direita.sety(y)

def raquete_direita_baixo():
    y = raquete_direita.ycor()
    y -= 30
    raquete_direita.sety(y)

# Teclado
janela.listen()
janela.onkeypress(raquete_esquerda_cima, "w")
janela.onkeypress(raquete_esquerda_baixo, "s")
janela.onkeypress(raquete_direita_cima, "Up")
janela.onkeypress(raquete_direita_baixo, "Down")

# Loop principal do jogo
while True:
    janela.update()

    # Movimentação da bola
    bola.setx(bola.xcor() + bola.dx)
    bola.sety(bola.ycor() + bola.dy)

    # Colisão com a parede superior
    if bola.ycor() > 190:
        bola.sety(190)
        bola.dy *= -1

    # Colisão com a parede inferior
    if bola.ycor() < -190:
        bola.sety(-190)
        bola.dy *= -1

    # Colisão com a parede direita
    if bola.xcor() > 290:
        bola.goto(0, 0)
        bola.dy *= -1
        placar_esquerda += 1
        placar.clear()
        placar.write("Jogador 1: {}  Jogador 2: {}".format(placar_esquerda, placar_direita), align="center", font=("Courier", 20, "normal"))

    # Colisão com a parede esquerda
    if bola.xcor() < -290:
        bola.goto(0, 0)
        bola.dy *= -1
        placar_direita += 1
        placar.clear()
        placar.write("Jogador 1: {}  Jogador 2: {}".format(placar_esquerda, placar_direita), align="center", font=("Courier", 20, "normal"))

    # Colisão com a raquete direita
    if (bola.xcor() > 240 and bola.xcor() < 250) and (bola.ycor() < raquete_direita.ycor() + 60 and bola.ycor() > raquete_direita.ycor() - 60):
        bola.setx(240)
        bola.dx *= -1

    # Colisão com a raquete esquerda
    if (bola.xcor() < -240 and bola.xcor() > -250) and (bola.ycor() < raquete_esquerda.ycor() + 60 and bola.ycor() > raquete_esquerda.ycor() - 60):
        bola.setx(-240)
        bola.dx *= -1
        
# Comando para fechar a janela
janela.mainloop()
```
